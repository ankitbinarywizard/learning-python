```python
import random
import hangman_words

stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

lives = 6

chosen_word_list = list(random.choice(hangman_words.word_list))
display_list = []

print(''' 
 _                                             
| |                                            
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __  
| '_ \ / _` | '_ \ / _` | '_ ` _ \ / _` | '_ \ 
| | | | (_| | | | | (_| | | | | | | (_| | | | |
|_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                    __/ |                      
                   |___/    ''')
for letter in range(len(chosen_word_list)):
    display_list += "_"

while "_" in display_list and lives > 0:    
    guess_letter = input("Make a guess: ").lower()
    if guess_letter in display_list:
        print(f"You have already guessed {guess_letter}, guess another letter.")
    for i in range(0, len(chosen_word_list)):
        if guess_letter == chosen_word_list[i]:
            display_list[i] = guess_letter
        
    print(display_list)
    
    if guess_letter not in chosen_word_list:
        lives -= 1
        print(f"Oh you guessed incorrectly. The letter {guess_letter} is not in the word!")
    print(stages[lives])
if lives > 0:
    print("You won!")
else:
    print("You lost!")
```
