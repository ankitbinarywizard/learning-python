```python
from random import randrange

cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

logo = """
.------.            _     _            _    _            _    
|A_  _ |.          | |   | |          | |  (_)          | |   
|( \/ ).-----.     | |__ | | __ _  ___| | ___  __ _  ___| | __
| \  /|K /\  |     | '_ \| |/ _` |/ __| |/ / |/ _` |/ __| |/ /
|  \/ | /  \ |     | |_) | | (_| | (__|   <| | (_| | (__|   < 
`-----| \  / |     |_.__/|_|\__,_|\___|_|\_\ |\__,_|\___|_|\_\\
      |  \/ K|                            _/ |                
      `------'                           |__/           
"""
start = input("Do you want to play a cards game called blackjack? Type 'y' or 'n'. ").lower()
if start == "y":
    print(logo)
    def blackjack():
        print("Welcome to the the game of Blackjack!")
        player_card_1 = cards[randrange(11)]
        player_card_2 = cards[randrange(11)]
        comp_card_1 = cards[randrange(11)]
        comp_card_2 = cards[randrange(11)]
        player_hand = [player_card_1, player_card_2]
        comp_hand = [comp_card_1, comp_card_2]
        player_score = player_card_1 + player_card_2
        comp_score = comp_card_1 + comp_card_2
        game = "on"
        print(f"Your cards are {player_hand}, and your score is {player_score}.")
        print(f"Computer's first card is {comp_hand[0]}.")
        player_input = input("Do you want another cards? Type 'y' or 'n'. ").lower()
        while player_input == "y":
            player_new_card = cards[randrange(11)]
            player_hand.append(player_new_card)
            player_score += player_new_card
            if player_score > 21:
                if player_new_card == 11:
                    player_score -= 10
                    print(f"Your cards are {player_hand}, and your score is {player_score}.")
                    player_input = input("Do you want another cards? Type 'y' or 'n'. ").lower()
                elif 11 in player_hand[:-1]:
                    player_score -= 10
                    print(f"Your cards are {player_hand}, and your score is {player_score}.")
                    player_input = input("Do you want another cards? Type 'y' or 'n'. ").lower()
                else:            
                    print(f"Your cards are {player_hand}, and your score is {player_score}.")
                    print("Game over! Computer won.")
                    game = "over"
                    break
            else:
                print(f"Your cards are {player_hand}, and your score is {player_score}.")
                player_input = input("Do you want another cards? Type 'y' or 'n'.").lower()
                
        if game == "over":
            next_game = input("Do you want to go again? Type 'y' or 'n'.").lower()
            if next_game == "y":
                blackjack()
            else:
                exit()
        else:
            print(f"Computer's cards are {comp_hand}, and its score is {comp_score}.")
            while comp_score < player_score:
                comp_new_card = cards[randrange(11)]
                comp_hand.append(comp_new_card)
                comp_score += comp_new_card
                if comp_score > 21:
                    if comp_new_card == 11:
                        comp_score -= 10
                        print(f"Computer's cards are {comp_hand}, and its score is {comp_score}.")
                    elif 11 in comp_hand[:-1]:
                        comp_score -= 10
                        print(f"Computer's cards are {comp_hand}, and its score is {comp_score}.")
                    else:
                        print(f"Computer's cards are {comp_hand}, and its score is {comp_score}.")
                        print("Game over! You won.")
                        game = "over"
                        break
                else:
                    print(f"Computer's cards are {comp_hand}, and its score is {comp_score}.")
            if game == "over":
                next_game = input("Do you want to go again? Type 'y' or 'n'.").lower()
                if next_game == "y":
                    blackjack()
                else:
                    exit()
            else:
                if comp_score == player_score:
                    print("Game drawn! Scores are equal.")
                    next_game = input("Do you want to go again? Type 'y' or 'n'.").lower()
                    if next_game == "y":
                        blackjack()
                    else:
                        exit()
                else:
                    print(f"Computer's score of {comp_score} is more than player's score i.e., {player_score}. Computer won!")       
                    next_game = input("Do you want to go again? Type 'y' or 'n'.").lower()
                    if next_game == "y":
                        blackjack()
                    else:
                        exit()
    blackjack()
else:
    print("Okay! No game for you.")
```
