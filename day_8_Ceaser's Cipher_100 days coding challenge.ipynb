```python
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
print("""           
 ,adPPYba, ,adPPYYba,  ,adPPYba, ,adPPYba, ,adPPYYba, 8b,dPPYba,  
a8"     "" ""     `Y8 a8P_____88 I8[    "" ""     `Y8 88P'   "Y8  
8b         ,adPPPPP88 8PP"""""""  `"Y8ba,  ,adPPPPP88 88          
"8a,   ,aa 88,    ,88 "8b,   ,aa aa    ]8I 88,    ,88 88          
 `"Ybbd8"' `"8bbdP"Y8  `"Ybbd8"' `"YbbdP"' `"8bbdP"Y8 88   
 
           88             88                                 
           ""             88                                 
                          88                                 
 ,adPPYba, 88 8b,dPPYba,  88,dPPYba,   ,adPPYba, 8b,dPPYba,  
a8"     "" 88 88P'    "8a 88P'    "8a a8P_____88 88P'   "Y8  
8b         88 88       d8 88       88 8PP""""""" 88          
"8a,   ,aa 88 88b,   ,a8" 88       88 "8b,   ,aa 88          
 `"Ybbd8"' 88 88`YbbdP"'  88       88  `"Ybbd8"' 88          
              88                                             
              88           
""")

while True:
    direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n")
    text = input("Type your message:\n").lower()
    shift = int(input("Type the shift number:\n"))

    def caeser(text_par, shift_par, dir_par):
        new_list = []
        if shift_par > 25:
            shift_par %= 25
        if dir_par == 'encode':
            text_par_list = list(text_par)
            for letter in text_par_list:
                if letter not in alphabet:
                    new_list.append(letter)
                elif alphabet.index(letter) + shift_par > 25:
                    new_list.append(alphabet[alphabet.index(letter) + shift_par - 26])
                else:
                    new_list.append(alphabet[alphabet.index(letter) + shift_par])
            new_text = "".join(new_list)
            print(f"Your encoded text is {new_text}.")
        elif dir_par == 'decode':
            text_par_list = list(text_par)
            for letter in text_par_list:
                if letter not in alphabet:
                    new_list.append(letter)
                elif alphabet.index(letter) - shift_par < 0:
                    new_list.append(alphabet[alphabet.index(letter) - shift_par + 26])
                else:
                    new_list.append(alphabet[alphabet.index(letter) - shift_par])
            new_text = "".join(new_list)
            print(f"Your decoded text is {new_text}.")
        elif dir_par != 'encode' or dir_par != 'decode' or shift_par > 25:
            print("Please provide correct direction.")

    caeser(text, shift, direction)
    go_again = input("Do you want to go again? Type 'Yes' or 'No'.").lower()
    if go_again != 'yes':
        break        
```
